# Derivatives - Introduction

The derivative of a function is the slope of the tangent line to the curve at a given point. It is defined as the limit of the average rate of change of the function as the interval approaches zero. The derivative can be interpreted as the instantaneous rate of change of the function with respect to its variable.

## The delta method

The delta method uses the difference between two points to calculate the slope at a particular point.

Put it simply:

$$m = \frac{\Delta y}{\Delta x}$$

Let's use the delta method on the function $y = x^2 + 2x + 2$, which at the end of the process will look like this:

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y

x = np.linspace(-10, 10, 1000)
y = f(x)
m = (10.61-10)/(2.1-2)
b = 10.61-m*2.1
line_y = m*x + b


fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', zorder=3)
ax.grid(True, linestyle='--', alpha=0.5)
_ = ax.plot(x,y)
```

Let's say we want to calculate the slope at $x = 2$.

First, determine the value of $y$ is:

```python
f(2)
```

Let's call that point $P$, located at $(2, 10)$.

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
ax.grid(True, linestyle='--', alpha=0.5)
plt.scatter(2, 10, label=r'$P$')
_ = ax.plot(x, y)
plt.legend()
plt.show()
```

We now need a second point, $Q$, relatively close to $P$. Let's say $Q$ is at $x = 5$:

```python
f(5)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10, label=r'$P$')
plt.scatter(5, 37, c = 'orange', zorder=3, label=r'$Q$')
ax.grid(True, linestyle='--', alpha=0.5)
_ = ax.plot(x,y)
plt.legend()
plt.show()
```

Now let's calculate the slope of the line connecting $P$ and $Q$, denoted as $m$:

$$m = \frac{change\ in \ y}{change \ in \  x} = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} = \frac{37 - 10}{5 - 2} = \frac{27}{3} = 9$$

```python
m = (37 - 10)/(5 - 2)
m
```

To plot the line connecting $P$ and $Q$ (denotes as $b$), we can rearrange the equation of a line $y = mx + b$ to solve for $b$:

$$b = y - mx$$

```python
b = 37 - m*5
b
```

And now we can calculate the equation of the line connecting $P$ and $Q$:

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10, label=r'$P$')
plt.scatter(5, 37, c = 'orange', zorder=3, label=r'$Q$')
plt.ylim(-5, 150) # new
plt.plot(x, line_y, c='orange', label=r'$b$')
ax.grid(True, linestyle='--', alpha=0.5)
_ = ax.plot(x,y)
plt.legend()
plt.show()
```

The closer $Q$ is to $P$, the closer the slope $m$ comes to be the true tangent of the point $P$.

So let's move $Q$ closer to $P$ by changing the value of $x$ to $2.1$:

```python
f(2.1)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10, label=r'$P$')
plt.scatter(2.1, 10.61, c = 'orange', zorder=3, label=r'$Q$')
ax.grid(True, linestyle='--', alpha=0.5)
_ = ax.plot(x,y)
plt.legend()
plt.show()
```

As $\Delta x$ approaches $0$, it becomes clearer that (in this example) the slope $m$ at point $P$ is equal to $6$.

To demonstrate this, let's change $\Delta x$ to $0.000001$:

```python
delta_x = 0.000001
delta_x
```

The value of $\Delta x$ is read as $1 \times 10^{-6}$.

Let's now represent $P$ with its coordinates:

```python
x1 = 2
y1 = 10
```

We can rearrange $\Delta x = x_2 - x_1$ to calculate $x_2$ for $Q$ as $x_2 = x_1 + \Delta x$:

```python
x2 = x1 + delta_x
x2
```

And $y_2$ can be calculated as $y_2 = f(x_2)$:

```python
y2 = f(x2)
y2
```

To find the slope $m$:

```python
m = (y2 - y1)/(x2 - x1)
m
```

If $x = -1$ (keep in mind that $x$ is $x_1$), we can calculate the slope at that point:

```python
x1 = -1
y1 = f(x1)
y1
```

Point $P$ is now at $(-1, 1)$.

```python
x2 = x1 + delta_x
x2
```

```python
y2 = f(x2)
y2
```

Point $Q$ is now at $(-0.999999, 1.000000000001)$, extremely close to $P$.

An alternative way to calculate $y_2$ is $f(x + \Delta x)$:

```python
y2 = f(x1 + delta_x)
y2
```

```python
m = (y2 - y1)/(x2 - x1)
m
```

As $x_2$ becomes infinitely close to $x_1$, it becomes clear that (in this example) the slope $m$ atg $x = -1$ is equal to $0$:

```python
b = y2 - m*x2
b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(x1, y1, label=r'$P$')
plt.scatter(x2, y2, c = 'orange', zorder=3, label=r'$Q$')
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', label=r'$b$')
ax.grid(True, linestyle='--', alpha=0.5)
_ = ax.plot(x,y)
plt.legend()
plt.show()
```

So, as $Q$ becomes infinitely close to $P$ $x_2 - x_1$ approaches $0$; therefore, $\Delta x$ approaches $0$ which can be denoted as $\Delta x \to 0$.

Using the delta method, we've derived the definition of differentiation from the first principles.

The derivative of $y$, $dy$, with respect to $x$,$dx$, can be represented as:

$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$$

Expanding $\Delta y$ to $y_2 - y_1$:

$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{y_2 - y_1}{\Delta x}$$

And replacing $y_1$ with $f(x)$:

$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

```python
def diff_demo(my_f, my_x, my_delta):
    return (my_f(my_x + my_delta) - my_f(my_x)) / my_delta
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
    print(diff_demo(f, 2, delta))
```

```python
for delta in deltas:
    print(diff_demo(f, -1, delta))
```
